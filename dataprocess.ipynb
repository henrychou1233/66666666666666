```python
import numpy as np
from PIL import Image
import os

def progressive_noise_and_denoise(clean_image_path, good_image_path, output_folder="diffusion_steps", num_steps=50):
    """
    Simulates a progressive noise addition (forward diffusion) and a reverse "denoising" process.

    Args:
        clean_image_path (str): Path to the initial clean image.
        good_image_path (str): Path to the target "good" (clean) image for the reverse process.
                                (In a real diffusion model, this would be generated, not provided.)
        output_folder (str): Folder to save the intermediate images.
        num_steps (int): Number of steps for noise addition and denoising.
    """

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # --- Load Images ---
    try:
        clean_img = Image.open(clean_image_path).convert("RGB")
        good_img = Image.open(good_image_path).convert("RGB")
    except FileNotFoundError:
        print(f"Error: Make sure image paths are correct. Could not find {clean_image_path} or {good_image_path}")
        return
    except Exception as e:
        print(f"An error occurred while opening images: {e}")
        return

    clean_np = np.array(clean_img)
    good_np = np.array(good_img)

    # Resize good_img to match clean_img dimensions if they are different
    if clean_np.shape != good_np.shape:
        print(f"Warning: 'good' image dimensions ({good_np.shape}) do not match 'clean' image dimensions ({clean_np.shape}). Resizing 'good' image.")
        good_img = good_img.resize(clean_img.size)
        good_np = np.array(good_img)


    print(f"Starting diffusion simulation with {num_steps} steps.")

    # --- Forward Diffusion (Adding Noise) ---
    print("\n--- Forward Diffusion: Adding Noise ---")
    current_img_np = np.copy(clean_np).astype(np.float32)
    noise_strength_per_step = 1 / num_steps * 255 # Max pixel value for 8-bit images

    for i in range(num_steps + 1):
        alpha = i / num_steps  # Alpha goes from 0 (clean) to 1 (full noise)

        # Generate Gaussian noise with the same shape as the image
        noise = np.random.normal(loc=0, scale=noise_strength_per_step, size=clean_np.shape)

        # Blend the current image with noise based on alpha
        # This is a simplified way to add noise progressively
        # A more accurate diffusion process involves specific noise schedules
        noisy_np = (1 - alpha) * clean_np + alpha * noise * np.max(clean_np) / noise_strength_per_step # Scale noise to be visible

        # Clamp values to 0-255 and convert back to uint8
        noisy_img_display = Image.fromarray(np.clip(noisy_np, 0, 255).astype(np.uint8))
        noisy_img_display.save(os.path.join(output_folder, f"noisy_step_{i:04d}.png"))
        print(f"Saved noisy_step_{i:04d}.png (Noise level: {alpha:.2f})")

    # The last image is considered "pure noise" (for this simulation)
    pure_noise_img_path = os.path.join(output_folder, f"noisy_step_{num_steps:04d}.png")
    print(f"\n--- Forward Diffusion Complete. Full noise image saved as: {pure_noise_img_path}")


    # --- Reverse Diffusion (Denoising/Generating) ---
    print("\n--- Reverse Diffusion: Denoising ---")
    # Start from the "pure noise" image (conceptually)
    # In a real model, this would be a random noise tensor
    current_denoised_np = np.copy(np.clip(noisy_np, 0, 255)).astype(np.float32) # Start from the last noisy image

    for i in range(num_steps + 1):
        alpha = i / num_steps  # Alpha goes from 0 (noisy) to 1 (clean)

        # Simulate denoising by blending the pure noise image back towards the "good" image
        # This is a *simplification*. A true diffusion model learns to predict noise and remove it.
        denoised_np = (1 - alpha) * current_denoised_np + alpha * good_np

        # Clamp values to 0-255 and convert back to uint8
        denoised_img_display = Image.fromarray(np.clip(denoised_np, 0, 255).astype(np.uint8))
        denoised_img_display.save(os.path.join(output_folder, f"denoised_step_{i:04d}.png"))
        print(f"Saved denoised_step_{i:04d}.png (Denoise level: {alpha:.2f})")

    print(f"\n--- Reverse Diffusion Complete. Clean image saved as: {os.path.join(output_folder, f'denoised_step_{num_steps:04d}.png')}")


# --- Usage Example ---
if __name__ == "__main__":
    clean_photo_path = r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\btad\one\test\ko\0048.png"
    good_photo_path = r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\btad\one\test\good\0016.png"
    output_directory = "diffusion_output"

    progressive_noise_and_denoise(clean_photo_path, good_photo_path, output_directory, num_steps=100)
    print(f"\nAll intermediate steps are saved in the '{output_directory}' folder.")
```
