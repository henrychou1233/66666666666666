```python
import os
from PIL import Image
from collections import Counter

# 把下面路徑列表改成你的實際路徑
dataset_paths = {
    "mvtec3d_restructured": r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\mvtec3d_restructured",
    "visa":                r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\visa",
    "mvtecloco":           r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\mvtecloco",
    "mvtec3d":             r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\mvtec3d",
    "mvtec":               r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\mvtec",
    "mpdd":                r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\mpdd",
    "mad":                 r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\mad",
    "btad":                r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\btad",
}

def scan_dataset(path):
    sizes = Counter()
    for root, _, files in os.walk(path):
        for fname in files:
            if fname.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".tiff")):
                try:
                    img = Image.open(os.path.join(root, fname))
                    sizes[img.size] += 1  # img.size = (width, height)
                    img.close()
                except Exception as e:
                    print(f"無法開啟 {os.path.join(root, fname)}：{e}")
    return sizes

if __name__ == "__main__":
    for name, path in dataset_paths.items():
        print(f"\n=== {name} ===")
        if not os.path.isdir(path):
            print("資料夾不存在或路徑錯誤，請確認路徑是否正確。")
            continue
        sizes = scan_dataset(path)
        if not sizes:
            print("找不到任何影像檔。")
            continue
        # 列出最常見的 5 種解析度
        for (w, h), cnt in sizes.most_common(5):
            print(f"解析度 {w}×{h}：共 {cnt} 張")
        # 如果有很多不同解析度，也可以印出總共幾種
        print(f"總共 {len(sizes)} 種不同解析度")
```

```python
import os

def rename_images_to_png(folder_path):
    valid_ext = ['.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff']
    files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
    images = [f for f in files if os.path.splitext(f)[1].lower() in valid_ext]
    images.sort()

    for idx, filename in enumerate(images):
        old_path = os.path.join(folder_path, filename)
        new_name = f"{idx:03d}.png"  # 統一改成png副檔名
        new_path = os.path.join(folder_path, new_name)

        if filename != new_name:
            # 若 new_path 已存在，先刪除避免衝突
            if os.path.exists(new_path):
                os.remove(new_path)
            os.rename(old_path, new_path)
            print(f"Renamed: {filename} -> {new_name}")

def rename_in_all_subfolders(root_folder):
    for dirpath, dirnames, filenames in os.walk(root_folder):
        rename_images_to_png(dirpath)

if __name__ == "__main__":
    folders = [
        r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\medical\chest\test",
        r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\medical\chest\train"
    ]

    for folder in folders:
        rename_in_all_subfolders(folder)

    print("全部檔案重新命名且副檔名改為 .png 完成。")
```

```python
import os
from PIL import Image

def convert_and_rename_images(folder_path):
    # 取得資料夾內所有檔案（包含子目錄）
    all_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # 加入完整路徑
            all_files.append(os.path.join(root, file))
    
    # 只保留圖片檔案（常見副檔名）
    img_exts = {'.jpg', '.jpeg', '.bmp', '.tif', '.tiff', '.gif', '.webp', '.png'}
    img_files = [f for f in all_files if os.path.splitext(f)[1].lower() in img_exts]
    
    # 依檔名排序（避免亂序）
    img_files.sort()

    # 開始逐一轉換、命名
    for idx, img_path in enumerate(img_files):
        try:
            img = Image.open(img_path).convert('RGB')  # 轉成 RGB 模式以避免透明問題
            # 新檔名：000.png, 001.png, ...
            new_name = f"{idx:03d}.png"
            new_path = os.path.join(os.path.dirname(img_path), new_name)
            img.save(new_path, 'PNG')
            if new_path != img_path:
                os.remove(img_path)  # 刪除原始檔案
            print(f"轉換成功: {img_path} -> {new_path}")
        except Exception as e:
            print(f"錯誤處理檔案 {img_path}: {e}")

if __name__ == "__main__":
    folders = [
        r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\medical\skin\test",
        r"C:\Users\anywhere4090\Desktop\hsnu\DDAD-main\dataset\medical\skin\train"
    ]
    
    for folder in folders:
        print(f"開始處理資料夾: {folder}")
        convert_and_rename_images(folder)
        print(f"完成資料夾: {folder}\n")
```
